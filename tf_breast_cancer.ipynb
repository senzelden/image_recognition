```python
import matplotlib.pyplot as plt
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras import backend as K
from sklearn.datasets import load_breast_cancer
```

```python
X, y = load_breast_cancer(return_X_y=True)
print(X.shape, y.shape)
```

```python
K.clear_session()

m = Sequential([
        Dense(4, activation='relu', input_shape=(30, )),
        Dense(4, activation='relu'),
        Dense(1, activation='sigmoid')
])
```

```python
m.summary()
```

```python
epochs=200
m.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])  # builds a Compution Graph
hist = m.fit(X, y, epochs=epochs, validation_split=0.2) # <-- mini batch gradient descent
```

```python
m.evaluate(X, y, batch_size=100)
```

```python
plt.plot(range(epochs), hist.history['loss'])
plt.plot(range(epochs), hist.history['val_loss'])
```

```python
plt.plot(range(epochs), hist.history['accuracy'])
plt.plot(range(epochs), hist.history['val_accuracy'])
```
