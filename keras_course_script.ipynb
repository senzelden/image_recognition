```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### 1. Load the data

```python
X, y = make_moons(1000, noise=0.1, random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
df = pd.DataFrame(X,y)
```

```python
df.reset_index(inplace=True)
```

```python
df.rename(columns={"index":"y", 0:"X_0", 1:"X_1"}, inplace=True)
```

```python
sns.scatterplot("X_0", "X_1", data=df, hue='y')
```

### 2. Import Keras

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense  # <-- Fully Connected, Feed-Forward
from tensorflow.keras import backend as K
```

### 3. Define a model

```python
K.clear_session()

m = Sequential([  # a list of layers
        Dense(4, activation="sigmoid", input_shape=(2, )), # hidden layer, input layer
        Dense(4, activation="sigmoid"), # another hidden layer
        Dense(1, activation="sigmoid") # output layer 
])
```

```python
m.summary()
```

### 4. Train a model

```python
m.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])  # builds a Compution Graph
hist = m.fit(X, y, epochs=300, batch_size=100, validation_split=0.2) # <-- mini batch gradient descent
```

```python
m.evaluate(X, y, batch_size=100)  # put test data here!
```
